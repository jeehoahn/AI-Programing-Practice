```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = Image.open('Clouds.jpg') # 이미지 파일 열기
im.show()                     # 이미지 파일 보여주기
```

```python
box = (100, 100, 1000, 1000)  # 자를 부분 정하기
region = im.crop(box)         # 이미지 파일 자르기
region.show()
```

```python
small = im.resize((100, 100)) # 이미지 크기 조절
small.show()
```

```python
Right = im.rotate(270)        # 이미지 270도 왼쪽을 돌리기
Right.show()
```

```python
Mirror = im.transpose(Image.Transpose.FLIP_LEFT_RIGHT) # 이미지 좌우 바꾸기
Mirror.show()
```

```python
from PIL import ImageFilter
Filter = im.filter(ImageFilter.SMOOTH_MORE) # 이미지 더 선명하게 보이는 필터 추가
Filter.show()
```

```python
Mirror.save('Clouds_Mirror.jpg') # 좌우 대칭한 이미지 'Clouds_Mirror'의 이름으로 코드 파일의 경로에 저장
```

```python
im_np = np.array(im)
plt.imshow(im_np)
plt.show()
```
