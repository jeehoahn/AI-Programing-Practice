# op.gg 크롤링 코드
op.gg는 게임의 데이터를 가지고 가장 좋은 캐릭터들을 순서대로 나열해준다 

-------
### 생성하는 코드의 목표 :
- op.gg의 소스 코드 긁어오기
+ 배정 받은 역할 입력 받기
- 1티어, 0티어 챔피언 모두 추천(5개 이하이면 아래 티어 캐릭터들을 포함해서 5개 나열 할 때까지)
+ 챔피언에 해당하는 티어, 승률등의 데이터 나열
<br><br>
### 코드 설명 :
1. op.gg는 자바 스크립트를 사용하기 때문에 셀레니옴으로 소스 코드르 긁어오기 <br><br>
2. 셀레니옴으로 긁어온 후 BeautifulSoup로 사이트의 내용물을 긁어오기 <br><br>
3. 1티어와 0티어 캐릭터들이 몇 개인지 변수로 저장하는 함수 생성, 5개 이하면 5를 저장 <br><br>
4. 개수 만큼 챔피언의 이름, 순위, 티어, 승률을 저장하는 리스트를 만들어주는 함수 작성<br><br>
5. 포지션 입력 받기<br><br>
6. 포지션에 해당하는 op.gg url 긁어오기<br><br>
7. 함수들을 통해서 리스트 만들기<br><br>
8. while문으로 리스트 안에 있는 내용을 한 개씩 print

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("--incognito")
    

def top5_check(champion_tier_1, champion_tier_op):        # 검색할 챔피언 수 정하기
    count=len(champion_tier_1)+len(champion_tier_op)
    if count>5:
        return count
    else:
        return 5

def ranking_list(num): # 가져올 챔피언의 순위 리스트 정리
    list1=[]
    i=0
    while i<num:
        i+=1
        list1.append(i)
    return list1

def champion_name_list(num): # 가져올 챔피언의 이름을 리스트에 넣어서 보내기
    list1=[]
    i=0
    td=soup.find_all('td',class_='css-1im1udv e1u05mw02')
    for name in td:
        list1.append(name.select_one('strong').text)
        i+=1
        if i==num:
            return list1

def champion_tier_list(num): # 가져올 챔피언의 티어를 리스트에 넣어서 보내기
    list1=[]
    i=0
    td=soup.find_all('td',class_='css-1qn65js e1u05mw05')
    for tier in td:
        tier=list(tier)
        if tier[1]=='0':
            list1.append('OP')
        else:
            list1.append(tier[1])
        i+=1
        if i==num:
            return list1
        
def champion_winrate_list(num): # 가져올 챔피언의 승률을 리스트에 넣어서 보내기
    list1=[]
    i=0
    td=soup.find_all('td', class_='css-9aydzo e1tupkk21')
    for winrate in td:
        i+=1
        if i%3==1:
            list1.append(list(winrate)[0]+'%')
        if i==num*3:
            return list1




pos=input('베정받은 포지션을 입력하세요 : ')
url='https://www.op.gg/champions?region=global&tier=emerald_plus&position='

if pos=='탑' or pos=='ㅌ' or pos=='top' or pos=='Top' or pos=='TOP' or pos=='x' or pos=='xkq':
    driver = webdriver.Chrome(options=chrome_options)
    url=url+'top'
    driver.get(url)
    page_source=driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    champion_tier_op=soup.find_all('span', class_='css-1dcegvi e1u05mw08')
    champion_tier_1=soup.find_all('span', class_='css-1q8h71u e1u05mw08')

    num=top5_check(champion_tier_1, champion_tier_op)
    ranking=ranking_list(num)
    name=champion_name_list(num)
    tier=champion_tier_list(num)
    winrate=champion_winrate_list(num)

    i=0
    while i<num:
        print(f'{tier[i]} Tier {ranking[i]}. {name[i]} Winrate : {winrate[i]}')
        i+=1

if pos=='정글' or pos=='ㅈㄱ' or pos=='wr' or pos=='wjdrmf' or pos=='Jungle' or pos=='jungle' or pos=='JUNGLE':
    driver = webdriver.Chrome(options=chrome_options)
    url=url+'jungle'
    driver.get(url)
    page_source=driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    champion_tier_op=soup.find_all('span', class_='css-1dcegvi e1u05mw08')
    champion_tier_1=soup.find_all('span', class_='css-1q8h71u e1u05mw08')

    num=top5_check(champion_tier_1, champion_tier_op)
    ranking=ranking_list(num)
    name=champion_name_list(num)
    tier=champion_tier_list(num)
    winrate=champion_winrate_list(num)

    i=0
    while i<num:
        print(f'{tier[i]} Tier {ranking[i]}. {name[i]} Winrate : {winrate[i]}')
        i+=1

if pos=='미드' or pos=='ㅁㄷ' or pos=='mid' or pos=='Mid' or pos=='MID' or pos=='ae' or pos=='alem':
    driver = webdriver.Chrome(options=chrome_options)
    url=url+'mid'
    driver.get(url)
    page_source=driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    champion_tier_op=soup.find_all('span', class_='css-1dcegvi e1u05mw08')
    champion_tier_1=soup.find_all('span', class_='css-1q8h71u e1u05mw08')

    num=top5_check(champion_tier_1, champion_tier_op)
    ranking=ranking_list(num)
    name=champion_name_list(num)
    tier=champion_tier_list(num)
    winrate=champion_winrate_list(num)

    i=0
    while i<num:
        print(f'{tier[i]} Tier {ranking[i]}. {name[i]} Winrate : {winrate[i]}')
        i+=1

if pos=='원딜' or pos=='ㅇㄷ' or pos=='adc' or pos=='Adc' or pos=='ADC' or pos=='de' or pos=='dnjself':
    driver = webdriver.Chrome(options=chrome_options)
    url=url+'adc'
    driver.get(url)
    page_source=driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    champion_tier_op=soup.find_all('span', class_='css-1dcegvi e1u05mw08')
    champion_tier_1=soup.find_all('span', class_='css-1q8h71u e1u05mw08')

    num=top5_check(champion_tier_1, champion_tier_op)
    ranking=ranking_list(num)
    name=champion_name_list(num)
    tier=champion_tier_list(num)
    winrate=champion_winrate_list(num)

    i=0
    while i<num:
        print(f'{tier[i]} Tier {ranking[i]}. {name[i]} Winrate : {winrate[i]}')
        i+=1

if pos=='서폿' or pos=='ㅅㅍ' or pos=='sup' or pos=='Sup' or pos=='SUP' or pos=='tv' or pos=='tjvht':
    driver = webdriver.Chrome(options=chrome_options)
    url=url+'support'
    driver.get(url)
    page_source=driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    champion_tier_op=soup.find_all('span', class_='css-1dcegvi e1u05mw08')
    champion_tier_1=soup.find_all('span', class_='css-1q8h71u e1u05mw08')

    num=top5_check(champion_tier_1, champion_tier_op)
    ranking=ranking_list(num)
    name=champion_name_list(num)
    tier=champion_tier_list(num)
    winrate=champion_winrate_list(num)

    i=0
    while i<num:
        print(f'{tier[i]} Tier {ranking[i]}. {name[i]} Winrate : {winrate[i]}')
        i+=1



driver.quit()
```
