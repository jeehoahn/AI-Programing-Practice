```python
#똥피지지

#자기가 배정된 포지션 입력하고
#티어 높은 챔피언 5개 만약 op, 1티어가 다섯개 이상이면 op 1티어 모두 추천

#자기가 선택한 챔피언을 치고
#가장 승률 높은 룬 세팅을 긁어오기

import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("--incognito")
    

def top5_check(champion_tier_1, champion_tier_op):        # 검색할 챔피언 수 정하기
    count=len(champion_tier_1)+len(champion_tier_op)
    if count>5:
        return count
    else:
        return 5

def ranking_list(num): # 가져올 챔피언의 순위 리스트 정리
    list1=[]
    i=0
    while i<num:
        i+=1
        list1.append(i)
    return list1

def champion_name_list(num): # 가져올 챔피언의 이름을 리스트에 넣어서 보내기
    list1=[]
    i=0
    td=soup.find_all('td',class_='css-1im1udv e1u05mw02')
    for name in td:
        list1.append(name.select_one('strong').text)
        i+=1
        if i==num:
            return list1

def champion_tier_list(num): # 가져올 챔피언의 티어를 리스트에 넣어서 보내기
    list1=[]
    i=0
    td=soup.find_all('td',class_='css-1qn65js e1u05mw05')
    for tier in td:
        tier=list(tier)
        if tier[1]=='0':
            list1.append('OP')
        else:
            list1.append(tier[1])
        i+=1
        if i==num:
            return list1
        
def champion_winrate_list(num):
    list1=[]
    i=0
    td=soup.find_all('td', class_='css-9aydzo e1tupkk21')
    for winrate in td:
        i+=1
        if i%3==1:
            list1.append(list(winrate)[0]+'%')
        if i==num*3:
            return list1




pos=input('베정받은 포지션을 입력하세요 : ')
url='https://www.op.gg/champions?region=global&tier=emerald_plus&position='

if pos=='탑' or pos=='ㅌ' or pos=='top' or pos=='Top' or pos=='TOP' or pos=='x' or pos=='xkq':
    driver = webdriver.Chrome(options=chrome_options)
    url=url+'top'
    driver.get(url)
    page_source=driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    champion_tier_op=soup.find_all('span', class_='css-1dcegvi e1u05mw08')
    champion_tier_1=soup.find_all('span', class_='css-1q8h71u e1u05mw08')

    num=top5_check(champion_tier_1, champion_tier_op)
    ranking=ranking_list(num)
    name=champion_name_list(num)
    tier=champion_tier_list(num)
    winrate=champion_winrate_list(num)

    i=0
    while i<num:
        print(f'{tier[i]} Tier {ranking[i]}. {name[i]} Winrate : {winrate[i]}')
        i+=1

if pos=='정글' or pos=='ㅈㄱ' or pos=='wr' or pos=='wjdrmf' or pos=='Jungle' or pos=='jungle' or pos=='JUNGLE':
    driver = webdriver.Chrome(options=chrome_options)
    url=url+'jungle'
    driver.get(url)
    page_source=driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    champion_tier_op=soup.find_all('span', class_='css-1dcegvi e1u05mw08')
    champion_tier_1=soup.find_all('span', class_='css-1q8h71u e1u05mw08')

    num=top5_check(champion_tier_1, champion_tier_op)
    ranking=ranking_list(num)
    name=champion_name_list(num)
    tier=champion_tier_list(num)
    winrate=champion_winrate_list(num)

    i=0
    while i<num:
        print(f'{tier[i]} Tier {ranking[i]}. {name[i]} Winrate : {winrate[i]}')
        i+=1

if pos=='미드' or pos=='ㅁㄷ' or pos=='mid' or pos=='Mid' or pos=='MID' or pos=='ae' or pos=='alem':
    driver = webdriver.Chrome(options=chrome_options)
    url=url+'mid'
    driver.get(url)
    page_source=driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    champion_tier_op=soup.find_all('span', class_='css-1dcegvi e1u05mw08')
    champion_tier_1=soup.find_all('span', class_='css-1q8h71u e1u05mw08')

    num=top5_check(champion_tier_1, champion_tier_op)
    ranking=ranking_list(num)
    name=champion_name_list(num)
    tier=champion_tier_list(num)
    winrate=champion_winrate_list(num)

    i=0
    while i<num:
        print(f'{tier[i]} Tier {ranking[i]}. {name[i]} Winrate : {winrate[i]}')
        i+=1

if pos=='원딜' or pos=='ㅇㄷ' or pos=='adc' or pos=='Adc' or pos=='ADC' or pos=='de' or pos=='dnjself':
    driver = webdriver.Chrome(options=chrome_options)
    url=url+'adc'
    driver.get(url)
    page_source=driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    champion_tier_op=soup.find_all('span', class_='css-1dcegvi e1u05mw08')
    champion_tier_1=soup.find_all('span', class_='css-1q8h71u e1u05mw08')

    num=top5_check(champion_tier_1, champion_tier_op)
    ranking=ranking_list(num)
    name=champion_name_list(num)
    tier=champion_tier_list(num)
    winrate=champion_winrate_list(num)

    i=0
    while i<num:
        print(f'{tier[i]} Tier {ranking[i]}. {name[i]} Winrate : {winrate[i]}')
        i+=1

if pos=='서폿' or pos=='ㅅㅍ' or pos=='sup' or pos=='Sup' or pos=='SUP' or pos=='tv' or pos=='tjvht':
    driver = webdriver.Chrome(options=chrome_options)
    url=url+'support'
    driver.get(url)
    page_source=driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    champion_tier_op=soup.find_all('span', class_='css-1dcegvi e1u05mw08')
    champion_tier_1=soup.find_all('span', class_='css-1q8h71u e1u05mw08')

    num=top5_check(champion_tier_1, champion_tier_op)
    ranking=ranking_list(num)
    name=champion_name_list(num)
    tier=champion_tier_list(num)
    winrate=champion_winrate_list(num)

    i=0
    while i<num:
        print(f'{tier[i]} Tier {ranking[i]}. {name[i]} Winrate : {winrate[i]}')
        i+=1



driver.quit()
```
